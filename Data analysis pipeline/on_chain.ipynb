```python
df_tx_count = pd.DataFrame(query_result_TX.result.rows)
df_tx_count.dtypes
```

```python
df_tx_count['block_date'] = pd.to_datetime(df_tx_count['block_date'], errors='coerce').dt.date
df_tx_count = df_tx_count.sort_values(by='block_date').reset_index(drop=True) 
```

```python
start_date = pd.to_datetime("2024-07-01").date()
end_date = pd.to_datetime("2025-04-10").date()

df_tx_count = df_tx_count[(df_tx_count['block_date'] >= start_date) & (df_tx_count['block_date'] <= end_date)] 
df_tx_count = df_tx_count.reset_index(drop=True)
```

```python
df_tx_count['block_date'] = pd.to_datetime(df_tx_count['block_date'], errors='coerce')
```

```python
onchain_df = pd.merge(df_add_vol, df_tx_count, left_on='time', right_on='block_date', how='inner')
```

```python
onchain_df.drop(columns=['block_date'], inplace=True)
onchain_df.rename(columns={'time': 'date'}, inplace=True)
```

```python
onchain_df.set_index('date', inplace=True)
```

```python
root_dir = os.path.join("..","final_datasets")
```

```python
onchain_df.to_csv(os.path.join(root_dir,"BTC_on_chain_metrics.csv"), index=True)
```

```python
import requests

url = "https://pro-api.coinmarketcap.com/v3/fear-and-greed/historical"
api_key = ""

headers = {
    "Accepts": "application/json",
    "X-CMC_PRO_API_KEY": api_key,
}

params = {
    "start": 1,  # optional: start date
    "interval": "daily",  # optional: interval of data
}

response = requests.get(url, headers=headers, params=params)

if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print("Error:", response.status_code, response.text)
```
