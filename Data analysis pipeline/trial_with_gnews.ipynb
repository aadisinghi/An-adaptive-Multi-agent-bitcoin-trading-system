```python
import requests
import json
from datetime import datetime, timedelta


query = 'crypto OR bitcoin OR btc OR cryptocurrencies OR blockchain'
max = '100'
expand = 'content'
sortby = 'relevance'
in_doc = 'title,description'

TRUSTED_SOURCES = {
    'CNBC', 'Yahoo Finance', 'Forbes', 'Business Insider',
    'Seeking Alpha', 'Fortune'
}

start_date = datetime.strptime('2024-07-01', '%Y-%m-%d')
end_date = datetime.strptime('2024-08-01', '%Y-%m-%d')

all_days_data = []

current_date = start_date
while current_date <= end_date:
    from_date = current_date.strftime('%Y-%m-%dT00:00:00Z')
    to_date = current_date.strftime('%Y-%m-%dT23:59:59Z')

    url = (
        f'https://gnews.io/api/v4/search?q={query}'
        f'&from={from_date}&to={to_date}'
        f'&in={in_doc}&sortby={sortby}&expand={expand}'
        f'&lang=en&max={max}&apikey={API_KEY}'
    )

    try:
        response = requests.get(url)
        data = response.json()
    except Exception as e:
        print(f"Error fetching data for {current_date.date()}: {e}")
        current_date += timedelta(days=1)
        continue

    # Deduplication for this day
    unique_keys = set()
    deduped_articles = []

    for article in data.get('articles', []):
        title = article.get('title', '').strip().lower()
        published_at = article.get('publishedAt', '').strip()
        key = f"{title}_{published_at}"

        if key not in unique_keys:
            source = article.get('source', {})
            source_name = source.get('name', '')
            if source_name in TRUSTED_SOURCES:
                deduped_articles.append({
                    "title": article.get('title', ''),
                    "publishedAt": article.get('publishedAt', ''),
                    "url": article.get('url', '')
                })
                unique_keys.add(key)

    # Store results
    all_days_data.append({
        "date": current_date.strftime('%Y-%m-%d'),
        "articles": deduped_articles
    })

    print(f"Fetched {len(deduped_articles)} trusted articles for {current_date.date()}")
    current_date += timedelta(days=1)

# Save to JSON file
with open("daily_crypto_news_july_august_2024.json", "w", encoding='utf-8') as f:
    json.dump(all_days_data, f, indent=2)

print("Saved all articles to daily_crypto_news_july_august_2024.json")
```
